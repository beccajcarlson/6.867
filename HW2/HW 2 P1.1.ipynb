```python
def sigmoid(z):
    """
    Return Sigmoid of z
    Args:
        Z : array or scalar
    """
    return 1.0 / (1.0 + np.exp(-z))
```

```python
def log_likelihood(features, target, weights, lamda):
    scores = np.dot(features, weights)
    ll = np.sum( target*scores - np.log(1 + np.exp(scores)) ) + lamda*(sum(np.square(weights[1:weights.shape[0]])))
    return ll
```

```python
def logistic_regression(features, target, num_steps, learning_rate, lamda, add_intercept = False):
    
    if add_intercept:
        
        intercept = np.ones((features.shape[0], 1))
        features = np.hstack((intercept, features))
        
    weights = np.zeros(features.shape[1])[np.newaxis].T
    target[target == -1] = 0
    
    intvec = []
    wt1vec = []
    wt2vec = []
    
    for step in xrange(num_steps):
        scores = np.dot(features, weights)
        predictions = sigmoid(scores)

        # Update weights with gradient
        output_error_signal = target - predictions
        gradient = np.dot(features.T, output_error_signal) - lamda*np.vstack((0, 2*weights[1:weights.shape[0]]))
        intvec.append(weights[0])
        wt1vec.append(weights[1])
        wt2vec.append(weights[2])
        weights = weights + learning_rate * gradient

        
        # Print log-likelihood every so often
        if step % 30000 == 0:
            print log_likelihood(features, target, weights, lamda)
        
    return weights, intvec, wt1vec, wt2vec
```
