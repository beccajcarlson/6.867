```python
# 2.1
def grad_descent(X, Y, W_init, w0_init, alpha, thresh, loss, f_prime, lamda):
    n = X.shape[0] #number of training examples
    n_iter = 1000000 # number of iterations  
    Xint = np.vstack((np.ones((X.shape[1], 1)).T, X))
    theta = W_init
    theta0 = w0_init
    prev_cost = loss(theta, theta0, Xint, Y, lamda)
#     next = np.vstack((prevint, prev)) - alpha*f_prime(prev, prevint, Xint, Y, lamda)[np.newaxis].T    
#     nextint = next[0][0]
#     next = next[0][1:3]
    for iteration in xrange(n_iter):
#         prev = next
#         prevint = nextint
        theta = np.vstack((theta0, theta)) - alpha*f_prime(theta, theta0, Xint, Y, lamda)[np.newaxis].T    
        theta0 = theta[0][0]
        theta = theta[0][1:3]
        curr_cost = loss(theta, theta0, Xint, Y, lamda)
        print curr_cost
        if abs(curr_cost - prev_cost) <= thresh:
            print('SUCCESS')
            return (iteration, curr_cost, theta0, theta)
        prev_cost = curr_cost
        
    print("Did Not Converge")
    return (iteration, curr_cost, theta0, theta)
```

```python
# Generate data on best lambda for validation - l2
accuracyl2val = []
accuracyl2test = []
intl2 = []
w2l2 = []
w1l2 = []

for data_num in [1, 2, 3, 4]:
    X, Y = data(data_num, 'train')
    Xval, Yval = data(data_num, 'validate')
    Xtest, Ytest = data(data_num, 'test')

    
    for lamda in [.01, 1, 12, 40]:
        clfl2 = sk.linear_model.LogisticRegression(penalty = 'l2', tol = 1e-10, C = 1/lamda, 
                                              fit_intercept = True, random_state = 10, 
                                         max_iter = 1000000, intercept_scaling = 1e6)
        clfl2.fit(X.T, np.ndarray.flatten(Y).T)
        accuracyl2val.append(sum((clfl2.predict(Xval.T) == Yval)[0])/len(Yval[0])*100.0000)
        accuracyl2test.append(sum((clfl2.predict(Xtest.T) == Ytest)[0])/len(Ytest[0])*100.0000)
        intl2.append(clfl2.intercept_[0])
        w2l2.append(clfl2.coef_.T[0][0])
        w1l2.append(clfl2.coef_.T[1][0])
    
    
accuracyl2test
```
